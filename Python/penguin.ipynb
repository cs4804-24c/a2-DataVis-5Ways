Penguin scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Penguins dataset
df = pd.read_csv('penglings.csv')

# Scatter plot
fig, ax = plt.subplots(figsize=(10, 6))

# Color mapping for the species
colors = {'Adelie': 'blue', 'Chinstrap': 'orange', 'Gentoo': 'green'}
df['SpeciesColor'] = df['species'].map(colors)

# Size based on bill length
scatter = ax.scatter(
    x=df['flipper_length_mm'],
    y=df['body_mass_g'],
    c=df['SpeciesColor'],
    s=df['bill_length_mm'] * 2,  
    alpha=0.8,
    label='_nolegend_'
)

ax.set_xlabel('Flipper Length (mm)')
ax.set_ylabel('Body Mass (g)')
ax.set_title('Penguin Scatter Plot')

ax.set_xlim(160, 240)
ax.set_ylim(2500, 6500)

ax.set_xticks(range(160, 241, 10))
ax.set_yticks(range(2500, 6501, 500))

# Grid
ax.grid(True, linestyle='--', alpha=0.7)

# legend for Species
legend_labels_species = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=species) for species, color in colors.items()]

# legend for Bill Lengths 
legend_labels_bill_length = [
    Line2D([0], [0], marker='o', color='black', markerfacecolor='black', markersize=10, label=f'Bill Length: 40'),
    Line2D([0], [0], marker='o', color='black', markerfacecolor='black', markersize=15, label=f'Bill Length: 50')
]
ax.legend(handles=legend_labels_species + legend_labels_bill_length, loc='upper left', bbox_to_anchor=(1.02, 1), title='Species and Bill Length')

plt.show()
```
