```python
#final

selected_species=["Adelie", "Gentoo", "Chinstrap"]
filtered_penguins=penguins[penguins['species'].isin(selected_species)]

colors={"Adelie": "darkorange", "Gentoo": "cyan", "Chinstrap": "darkorchid"}

sns.set_style("whitegrid")
sns.set_palette("gray")


plt.figure(figsize=(10,6))

scatter=sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', size='bill_length_mm',
                       sizes=(10,100), palette=colors,data=filtered_penguins, alpha=0.8)

plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.legend(bbox_to_anchor=(1, 0.55), loc='upper right')
```
