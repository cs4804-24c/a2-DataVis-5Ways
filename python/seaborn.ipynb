```python
# %pip install pandas
# %pip install matplotlib
# %pip install seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csvFile = pd.read_csv('../penglings.csv').dropna()

colors = {
    'Adelie': '#f8a452',
    'Gentoo': '#4ea0a1',
    'Chinstrap': '#b056cf'
}

sns.set(style="whitegrid")

plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    size='bill_length_mm',
    sizes=(40, 300),
    alpha=0.80,
    palette=colors,
    data=csvFile,
    edgecolor='None'
)

plt.legend(loc='lower right')

plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')

plt.show()
```
