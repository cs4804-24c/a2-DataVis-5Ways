```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact, widgets
import altair as alt
```

```python
csv_file_path = "/Users/shivalimani/Library/CloudStorage/OneDrive-WorcesterPolytechnicInstitute(wpi.edu)/Senior Year/Data Visualization/a2-DataVis-5Ways/penglings.csv"
peng = pd.read_csv(csv_file_path)
peng
```

```python
def peng_scatterplot(size_mapping_active, color_mapping_active):
    sns.set_style("whitegrid")
    color_mapping = {'Adelie': 'orange', 'Chinstrap': 'purple', 'Gentoo': 'blue'}
    plt.xlabel("Flipper Length (mm)")
    plt.ylabel("Body Mass (g)")
    
    if size_mapping_active:
        size_value = 'bill_length_mm'
    else:
        size_value = None  # No size mapping

    if color_mapping_active:
        color_value = 'species'
    else:
        color_value = None  # No color mapping

    sns.scatterplot(x="flipper_length_mm", y="body_mass_g", data=peng, size=size_value, hue=color_value, palette=color_mapping, alpha=0.6)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
# Use interact to create an interactive UI
interact(peng_scatterplot, size_mapping_active=widgets.Checkbox(value=True, description='Activate Size Mapping'),
         color_mapping_active=widgets.Checkbox(value=True, description='Activate Color Mapping'))

# Show the interactive plot
plt.show()
```

```python
def altair_scatterplot(size_mapping_active, color_mapping_active):
    color_mapping = {'Adelie': 'orange', 'Chinstrap': 'purple', 'Gentoo': 'blue'}
    # Filter data based on user input
    if size_mapping_active:
        size_value = alt.Size('bill_length_mm:O', title='Size Mapping', scale=alt.Scale(domain=[35, 40, 45, 50, 55], range=[50, 100, 150, 200, 250]))
    else:
        size_value = alt.value(50)

    if color_mapping_active:
        color_value = alt.Color('species:N', scale=alt.Scale(domain=list(color_mapping.keys()), range=list(color_mapping.values())))
    else:
        color_value = alt.value('black')

    # Create Altair chart
    chart = alt.Chart(peng).mark_circle().encode(
        x=alt.X('flipper_length_mm:Q', title='Flipper Length (mm)', scale=alt.Scale(zero=False)),
        y=alt.Y('body_mass_g:Q', title='Body Mass (g)', scale=alt.Scale(zero=False)),
        size=size_value,
        color=color_value,
        tooltip=['species:N', 'flipper_length_mm:Q', 'body_mass_g:Q']
    ).properties(
        width=600,
        height=400
    ).configure_mark(opacity=0.6)  # Set opacity to 60%

    # Show the chart
    chart.display()

# Use interact to create an interactive UI
interact(altair_scatterplot, size_mapping_active=widgets.Checkbox(value=True, description='Activate Size Mapping'),
         color_mapping_active=widgets.Checkbox(value=True, description='Activate Color Mapping'))
```
